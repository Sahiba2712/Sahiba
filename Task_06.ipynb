 1. Load the necessary package for plotting using pyplot from matplotlib. Example - Days(x-axis) represents 8 days and Speed represents a carâ€™s speed. Plot a Basic line plot between days and car speed, put x axis label as days and y axis label as car speed and put title Car Speed Measurement.
 
         Days=[1,2,3,4,5,6,7,8]
        
 Speed=[60,62,61,58,56,57,46,63]

```python
import matplotlib.pyplot as plt
Days=[1,2,3,4,5,6,7,8]
Speed=[60,62,61,58,56,57,46,63]
plt.plot(Days,Speed)
plt.xlabel('Days')
plt.ylabel('Car speed')
plt.title('car speed')
plt.show()
```

   **2. Now to above car data apply some string formats  like line style example green dotted line, marker shape like +, change markersize, markerface color etc.**

```python
plt.plot(Days,Speed, 'go--', marker='+', linewidth=2, markersize=10)
```

 4. Plotting a basic sine graph by adding more features. Adding Multiple plots by Superimposition like cosine wave.

```python
import numpy as np
import math
a=np.arange(0,math.pi*2,0.05)
b=np.sin(a)
c=np.cos(a)
plt.plot(a,b,'r',label='sine')
plt.plot(a,c,'b',label='cosine')
plt.xlabel('Angle')
plt.ylabel('Magnitude')
plt.title('Superposition of cosine and sine wave')
plt.legend()
plt.show()
```

 5. Plot Simple bar chart showing popularity of Programming Languages.

Plot Multiple Bars showing Popularity and Security of major Programming Languages. Also Create Horizontal bar chart using barh function.

```python
 
L =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
 
P = [56, 39, 34, 34, 29]
 
S= [44 ,36 ,55, 50, 42]
plt.barh(L,P,color='r',label='Popularity')
plt.barh(L,S,color='g',label='Security')
plt.title('Programming languages')
plt.legend()
```

 6.  Plot Histogram, We have a sample data of Students marks of various Students, we will try to plot number of Students by marks range and try to figure out how many Students are average, below-average and Excellent.
 
Marks=[ 61,86,42,46,73,95,65,78,53,92,55,69,70,49,72,86,64]
 
Histogram showing Below Average, Average and Execellent distribution
 
40-60: Below Average
 
60-80: Average
 
80-100: Excellent

```python
 Marks=[ 61,86,42,46,73,95,65,78,53,92,55,69,70,49,72,86,64]
M2=[i for i in Marks if 40<i<60]
M3=[i for i in Marks if 60<i<80]
M4=[i for i in Marks if 60<i<100]
 
plt.hist(M2,bins=10,label='Below')
plt.hist(M3,label='average')
plt.hist(M4,label='Excellent')
plt.legend()
plt.show()
```
